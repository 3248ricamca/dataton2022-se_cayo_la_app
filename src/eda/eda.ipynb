# Noticias

```python
noticias = pd.read_csv("../datos/clientes_noticias.csv")
noticias['Portal_noticia'] = noticias['news_url_absolute'].apply(lambda x: x.split('/')[2])
noticias['categoria_noticia'] = noticias['news_url_absolute'].apply(lambda x: x.split('/')[3])
```

```python
portales = noticias.groupby('Portal_noticia').count()['nit'].reset_index()
portales['participacion'] = round(portales['nit']/portales['nit'].sum()*100,2)
portales.sort_values('nit', ascending = False, inplace = True)
portales['por_acumulado'] = portales['participacion'].cumsum()
portales.head(30)
```

```python
portales = noticias.groupby('categoria_noticia').count()['nit'].reset_index()
portales['participacion'] = round(portales['nit']/portales['nit'].sum()*100,2)
portales.sort_values('nit', ascending = False, inplace = True)
portales['por_acumulado'] = portales['participacion'].cumsum()
portales.head(30)
```
